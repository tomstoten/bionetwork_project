```python
import networkx as nx
from networkx.algorithms.bipartite.matrix import from_biadjacency_matrix
import cobra

from cobra.io import read_sbml_model
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def create_network(model):
    G = nx.Graph()

    for reaction in model.reactions:
        reaction_id = reaction.id
        reaction_name = reaction.name
        G.add_node(reaction_id, type="reaction", name=reaction_name)

        for metabolite in reaction.metabolites:
            species_id = metabolite.id
            species_name = metabolite.name
            species_formula = metabolite.formula  
            G.add_node(species_id, type="metabolite", name=species_name, formula=species_formula)
            G.add_edge(reaction_id, species_id)

    return G

```

```python
color_list = ['red', 'blue', 'green', 'yellow']

def draw_network(G, node_type_attribute='type'):
    pos = nx.spring_layout(G, seed=33)  # Adjust layout as needed (e.g., nx.kamada_kawai_layout)

    node_types = set(nx.get_node_attributes(G, node_type_attribute).values())

    # Create a dictionary to map node types to colors
    node_type_to_color = {node_type: color_list[i % len(color_list)] for i, node_type in enumerate(node_types)}

    for node_type, color in node_type_to_color.items():
        nodes_of_type = [n for n, d in G.nodes(data=True) if d[node_type_attribute] == node_type]
        nx.draw_networkx_nodes(G, pos, nodelist=nodes_of_type, node_size = 25, node_color=color, label=node_type)

    nx.draw_networkx_edges(G, pos, alpha=0.5)

    plt.legend()
    plt.axis('off')
    plt.show()
```

```python
def NetworkFromSBML(model):
    
    G = nx.Graph()
    
    # add metabolite nodes to graph
    for met in model.metabolites:
        G.add_node(met.id, bipartite = 0)
        
        
    for rxn in model.reactions:
        G.add_node(rxn.id, bipartite = 1)
        for met in rxn.metabolites:
            G.add_edge(rxn.id, met.id)
    
    #pos = bipartite_layout(G, G.nodes())
    
    return G
    
```

```python
model = read_sbml_model("sbml_files/protein.sbml")
```

```python
mycoplasma_network = create_network(model)
```

```python
draw_network(mycoplasma_network)
```
